```python
import numpy as np
import pyttsx3  
import pandas as pd
import PySimpleGUI as pg  
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv('crop_detection.csv')

#engine = pyttsx3.init('sapi5')                                            
#voices = engine.getProperty('voices')
#rate = engine.getProperty('rate')
#engine.setProperty('rate', rate-20)
#engine.setProperty('voice',voices[1].id)

#def speak(audio):                                                         
#    engine.say(audio) 
#    engine.runAndWait()

x = df.drop('CROP', axis=1)
y = df['CROP']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
prediction = classifier.predict(x_test)
accuracy = accuracy_score(y_test, prediction)
print(accuracy*100, "%")

layout = [[pg.Text('                      Crop Recommendation Assistant', font=("Helvetica", 30), text_color = 'yellow')],                                                   
         [pg.Text('Enter ratio of Nitrogen in the soil                                  :', font=("Helvetica", 20)), pg.Input(font=("Helvetica",20), size = (20,1) )],
         [pg.Text('Enter ratio of Phosphorous in the soil                           :', font=("Helvetica", 20)), pg.Input(font=("Helvetica", 20),size = (20,1))],
         [pg.Text('Enter ratio of Potassium in the soil                               :', font=("Helvetica", 20)), pg.Input(font=("Helvetica", 20),size = (20,1))],
         [pg.Text('Enter average Temperature value around the field        :', font=("Helvetica", 20)), pg.Input(font=("Helvetica", 20),size = (20,1)), pg.Text('*C', font=("Helvetica", 20))], 
         [pg.Text('Enter average percentage of Humidity around the field :', font=("Helvetica", 20)), pg.Input(font=("Helvetica", 20),size = (20,1)), pg.Text('%', font=("Helvetica", 20))], 
         [pg.Text('Enter PH value of the soil                                            :', font=("Helvetica", 20)), pg.Input(font=("Helvetica", 20),size = (20,1))], 
         [pg.Text('Enter average amount of Rainfall around the field        :', font=("Helvetica", 20) ), pg.Input(font=("Helvetica", 20),size = (20,1)),pg.Text('mm', font=("Helvetica", 20))],
         [pg.Text(size=(50,1),font=("Helvetica",20) , text_color = 'yellow', key='-OUTPUT1-' )],
         [pg.Button('Submit', font=("Helvetica", 20)),pg.Button('Quit', font=("Helvetica", 20))] ]
window = pg.Window('Crop Recommendation Assistant', layout) 

while True: 
    event, values = window.read()
    if event == pg.WINDOW_CLOSED or event == 'Quit':                                                                                            
        break
    nitrogen_content = values[0]                                                                                                    
    phosphorus_content = values[1]                                                                                                        
    potassium_content = values[2]                                                                                                       
    temperature_content = values[3]                                                                                                        
    humidity_content = values[4]                                                                                                        
    ph_content = values[5]                                                                                                        
    rainfall = values[6]
    input_data = ([nitrogen_content,phosphorus_content, potassium_content, temperature_content,
                   humidity_content, ph_content, rainfall])
    input_data_as_numpy = np.asarray(input_data)
    input_data_reshaped = input_data_as_numpy.reshape(1, -1)
    pred = classifier.predict(input_data_reshaped)
    
    if int(humidity_content)<= 33 :                                                
        humidity_level = 'low humid'
    elif int(humidity_content) >=34 and int(humidity_content) <= 66:
        humidity_level = 'medium humid'
    else:
        humidity_level = 'high humid'

    if int(temperature_content)<= 6:                                           
        temperature_level = 'cool'
    elif int(temperature_content) >=7 and int(temperature_content) <= 25:
        temperature_level = 'warm'
    else:
        temperature_level= 'hot' 
    
    if int(rainfall) <= 100:                                                              
        rainfall_level = 'less'
    elif int(rainfall) >= 101 and int(rainfall) <=200:
        rainfall_level = 'moderate'
    else:
        rainfall_level = 'heavy rain'

    if int(nitrogen_content) <= 50:                                             
        nitrogen_level = 'less'
    elif int(nitrogen_content) >=51 and int(nitrogen_content) <=100:
        nitrogen_level = 'not to less but also not to high'
    else:
        nitrogen_level = 'high'

    if int(phosphorus_content) <= 50:                                         
        phosphorus_level = 'less'
    elif int(phosphorus_content) >= 51 and int(phosphorus_content) <=100:
        phosphorus_level = 'not to less but also not to high'
    else:
        phosphorus_level = 'high'

    if int(potassium_content) <=50:                                         
        potassium_level = 'less'
    elif int(potassium_content) >= 51 and int(potassium_content) <= 100:
        potassium_level = 'not to less but also not to high'
    else:
        potassium_level = 'high'

    if float(ph_content) >=0 and float(ph_content) <=5:                                                        
        phlevel = 'acidic' 
    elif float(ph_content) >= 6 and float(ph_content) <= 8:
        phlevel = 'neutral'
    elif float(ph_content) >= 9 and float(ph_content) <= 14:
        phlevel = 'alkaline'
    else:
        phlevel = 'PH level not detected'

    print(humidity_level)
    print(temperature_level)
    print(rainfall_level)
    print(nitrogen_level)
    print(phosphorus_level)
    print(potassium_level)
    print(phlevel)
    
    #speak("Sir according to the data that you provided to me. The ratio of nitrogen in the soil is  " + nitrogen_level + ". The ratio of phosphorus in the soil is  " + phosphorus_level + ". The ratio of potassium in the soil is  " + potassium_level + ". The temperature level around the field is  " + temperature_level +". The humidity level around the field is  " + humidity_level + ". The ph type of the soil is  " + phlevel + ". The amount of rainfall is  " + rainfall_level) 
    window['-OUTPUT1-'].update('The best crop that you can grow : ' + pred)
    #speak("The best crop that you can grow is  " + pred) 
window.close()
```
